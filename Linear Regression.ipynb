# Write linear regression model to predict the profit of 50 startups.

### Step 1: Import Libraries


```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

### Step 2: Data Preperations

#### 2.1 Reading Excel file

```python
sampledata = pd.read_csv('50_Startups.csv',header=0)
```

```python
sampledata
```

#### 2.2 Spliting Dataset (40 Instances for Training / 10 Instances for Testing)

```python
x = sampledata [['R&D Spend','Administration', 'Marketing Spend']].values
y = sampledata ['Profit'].values
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
print (x_train)
print ("-"*100)
print (y_train)
print ("-"*100)
print(x_test)
print ("-"*100)
print(y_test)
```

### Step 3: Data Scaling

```python
def data_scaling(x):
    x_scaled_value = x-np.mean(x,axis=0)
    x_scaled_value = x_scaled_value / np.std(x_scaled_value,axis=0)
    return x_scaled_value
```

```python
x_train_scaling = data_scaling(x_train)
x_test_scaling = data_scaling(x_test)
print(x_train_scaling)
print("-"*100)
print(x_test_scaling)
```

### Step 4: Running Gradiant Decent Function

```python
# Function to fit the gradiant decent
def fitingGradientDecent(x, y, iterations, LR):
    cost = []
        
    w = np.zeros((1 + x.shape[1]))
    m = x.shape[0]
    
    for i in range(iterations):
        y_pred = np.dot(x,w[1:]) + w[0]
        mse = (1/m) * np.sum((y_pred - y)**2)
        cost.append(mse)
            
        #Derivatives
        d_w1 = (2/m) * np.dot(x.T, (y_pred - y))
        d_w0 = (2/m) * np.sum(y_pred - y)
            
        #Values update
        w[1:] = w[1:] - LR * d_w1
        w[0] = w[0] - LR * d_w0
    return w,cost
```

```python
def predictGradientDecent(x,w):
        return w[0] + np.dot(x, w[1:])
```

### Step 5: Initalizing learning rate and Iterations

```python
Learning_rate = 0.1
Iterations = 100
```

### Step 6: Calculating w and cost

```python
w_values,cost_values=fitingGradientDecent(x_train_scaling,y_train,Iterations, Learning_rate)
```

```python
print("w0 = ", w_values[0])
print("w1 = ", w_values[1])
print("w2 = ", w_values[2])
print("w3 = ", w_values[3])
```

```python
print(cost_values)
```

### Step 7: Model Learning
##### 1- Predict Y values bases on new weight and training data set

```python
y_pred = predictGradientDecent(x_train_scaling,w_values)
print(y_pred)
```

##### 2- Calculating Error 

```python
errors = np.sum((y_pred - y_train)**2)
sst = np.sum((y_train - np.mean(y_train))**2)
R_square = 1 - (errors/sst)

R_square_adjusted = 1 - ((1 - R_square) * (x_train.shape[0] - 1) / (x_train.shape[0] - x_train.shape[1] - 1))
print(R_square_adjusted)
```

##### 3- Comparing Initial vs Final Cost according to Iteration number

```python
plt.title('Cost Values')
plt.plot(cost_values)
plt.ylabel('Cost')
plt.xlabel('Iteration')
print('Initial cost\t', cost_values[0])
print('Final cost\t', cost_values[-1])
```
